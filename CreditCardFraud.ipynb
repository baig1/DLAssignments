```python
from sklearn.preprocessing import StandardScaler
df_norm = df
df_norm['Time'] = StandardScaler().fit_transform(df_norm['Time'].values.reshape(-1, 1))
df_norm['Amount'] = StandardScaler().fit_transform(df_norm['Amount'].values.reshape(-1, 1))
```

```python
train_data = df_norm.iloc[:199365,:]
train_data.shape
```

```python
test_data = df_norm.iloc[199365:,:]
test_data.shape
```

```python
train_data.head()
```

```python
test_data.head()
```

```python
train_data = train_data.sample(frac=1)
train_data.head()
```

```python
test_data = test_data.sample(frac = 1)
test_data.head()
```

```python
train_labels = train_data.pop('Fraud')
test_labels = test_data.pop('Fraud')
```

```python
train_labels
```

```python
test_labels
```

```python
model = Sequential()
model.add(Dense(10, activation = 'relu', input_shape = [len(train_data.keys())]))
model.add(Dense(8, activation = 'relu'))
model.add(Dense(6, activation = 'relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
model.summary()
```

```python
history = model.fit(train_data, train_labels, epochs = 100, batch_size = 30, shuffle=True, validation_split=0.2, verbose = 1,).history
```

```python
predictions = model.predict(test_data)
predictions = predictions.flatten()
```

```python
predictions
```

```python
testing = model.evaluate(test_data, test_labels)
```
